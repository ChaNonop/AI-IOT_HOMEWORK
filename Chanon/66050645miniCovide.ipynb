### Read data from csv


```python
def read_data(file_name):
    with open(file_name, mode='r', encoding ='utf-8-sig') as f:
        data = f.readlines()
        return data
```

### Transform data 

```python
def transform_data(data, keys):
    cases = list()
    case = dict()
    for item in data:
        record = item.strip().split(',')
        for i in range(len(keys)):
            case[keys[i]] = record[i]
        cases.append(case)
        case = dict()
    return cases
```

#### insight1

```python
def number_of_cases(cases, date):
    count = 0
    for case in cases:
        if case['announce_date'] == date:
            count += 1
    return count
```

### Insight 2

```python
#แยกเพศ แยกวันที่
def number_of_cases_by_sex(cases, start_date, end_date):
    males = 0
    females = 0
    unknown = 0
    for case in cases:
        if start_date <= case['announce_date'] <= end_date:
            if case['sex'] == 'ชาย' :
                males += 1
            elif case['sex'] == 'หญิง' :
                females += 1
            else:
                unknown += 1
    return males, females, unknown
```

```python
m,f,un = number_of_cases_by_sex(cases, '15/8/2021', '16/8/2021')
print("male ", m , "female" , f, 'unknown', un)
```

### Insight3

```python
#ดูการกระจายตัวตามช่วงอายุที่กําหนด
def number_of_cases_by_ages(cases,age_range):
    bin = [0]*(len(age_range)+1) #bin= 0,0,0,0,0,0,0

    for case in cases:
        if case['Unit'] == 'เดือน' or case['Unit'] == 'วัน' :
            bin[0] += 1
        elif case['Unit'] == 'ปี' :
            for i in range(len(bin)-1):
                if age_range[i][0] <= float(case['age']) <= age_range[i][1]:
                    bin[i] += 1
                    break
        else:
            bin[-1]+= 1
    return bin
```

```python
#ดูแยกตาม range อายุ
print (number_of_cases_by_ages(cases, [(0,19),(20,39), (40,59), (60,79), (80,99), (100, 119)]) )
```

### Insight 5

```python
# หาพื้นที่ที่มีความเสี่ยงสูงจากจังหวัดที่มีผู้ติดเชื้อมากที่สุด
def top_risk_provinces(cases, top_n=10):
    province_count = {}
    for case in cases:
        province = case.get('province_of_isolation')
        if province:
            province_count[province] = province_count.get(province, 0) + 1
        else:
            province_count['unknown'] = province_count.get('unknown', 0) + 1
    # เรียงลำดับจากมากไปน้อย
    sorted_prov = sorted(province_count.items(), key=lambda x: x[1], reverse=True)
    print(f"10 จังหวัดที่มีผู้ติดเชื้อสูงสุด")
    for prov, count in sorted_prov[:top_n]:
        print(f"{prov}: {count} ราย")

top_risk_provinces(cases, top_n=10)
```
